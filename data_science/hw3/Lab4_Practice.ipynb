```python
import pandas as pd
import numpy as np
import requests
date = "20230327"
url = f'https://www.twse.com.tw/exchangeReport/MI_INDEX?response=json&date={date}&type=ALLBUT0999'
response = requests.get(url)
response_json = response.json()
stock = pd.DataFrame(response_json['data9'], columns=response_json['fields9'])

# 收盤價
rec_st = stock[['證券名稱','收盤價']].copy()
rec_st
```

```python
# print(type(rec_st.收盤價[0]))
# print(type(rec_st['收盤價']>1000))
replace_map = {",":"", "--":np.nan}
rec_st = rec_st.replace(replace_map, regex=True)
rec_st.收盤價 = pd.to_numeric(rec_st.收盤價, errors='coerce', downcast='float')

data = rec_st[rec_st['收盤價']>1000] # find > 1000
print(data)
print(data.describe())
print('\n大於1000的公司數量 {:d}'.format(data['證券名稱'].count()))
print('\n平均收盤價 {:.1f}'.format(data['收盤價'].mean()))
print('\n股王是', data[data['收盤價'] == data['收盤價'].max()]['證券名稱'])
```
